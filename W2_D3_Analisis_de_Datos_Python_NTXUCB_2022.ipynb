## Librería Matplotlib

Es una libreria de visualización de datos (mayormente en 2 dimensiones) que hace visualizaciones similares a las que se realizan en MATLAB.

```python
%matplotlib inline
```

```python
# Importamos las librerias
import matplotlib.pyplot as plt # Libreria Matplotlib
import numpy as np
import pandas as pd
```

```python
# Forma más básica
data = np.random.rand(10)
data
```

```python
plt.plot(data)
```

### Scatter Plots

```python
x = np.random.randint(0,20,50)
y = np.random.randint(0,30,50)
plt.scatter(x,y)
plt.show()
```

### Histogramas

```python
data = np.random.normal(0,1,10000)
plt.hist(data,bins=100)
plt.show()
```

## Figuras y Subfiguras

```python
data = np.random.rand(15)
```

```python
fig = plt.figure()
fig.set_figheight(6)
fig.set_figwidth(10)
ax1 = fig.add_subplot(2,2,1)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Plot 1')
plt.plot(data)
ax2 = fig.add_subplot(2,2,2)
ax2.set_xlabel('Años')
ax2.set_ylabel('Densidad')
ax2.set_title('Densidad por año')
plt.plot(data,'go--')
ax3 = fig.add_subplot(2,2,3)
ax3.set_title('Histograma 1 con 3 bins')
plt.hist(data,bins=3)
ax4 = fig.add_subplot(2,2,4)
ax4.set_title('Histograma 2 con 5 bins')
plt.hist(data,bins=5)
fig.show()
```

```python
# Varios graficos en un mismo plot
arr1 = np.arange(-13,13)
arr2 = arr1 ** 2

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(arr1,'r:',label='x')
ax.plot(arr2,'b--',label='x^2')
ax.legend()
fig.show()
```

## Visualización usando Pandas

```python
## Importando datos desde una fuente disponible en la web
df = pd.read_csv('https://raw.githubusercontent.com/ntxucb/W2-Introduccion-Data-Science/main/datasets/population_total.csv')
df.head()
```

```python
## Preproceso
## Paso 1 quitar valores faltantes
df = df.dropna()
## Paso 2 ajustar la forma de ser necesario
df = df.pivot(index='year', columns='country',values='population')
df
```

### Line Plots

```python
## Ploteamos los datos de la población en Bolivia
df[['Bolivia','Ecuador']].plot(grid=True,style=['ro-','b^-'],ylabel='population',title='Bolivia vs Ecuador')
```

## Bar Plots

```python
df[['Argentina','Colombia','Bolivia']].plot.bar(figsize=(15,7))
```

```python
df[['Argentina','Colombia']].plot.barh()
```

```python
df[['Argentina','Colombia']].plot.barh(stacked=True)
```

```python
df.iloc[0].plot.bar(figsize=(20,7))
```

### Otros Graficos

```python
df.iloc[0].plot.pie()
```

```python
df[['Chile','Peru','Ecuador']].plot.box()
```

```python
df.iloc[0].plot.box()
```
