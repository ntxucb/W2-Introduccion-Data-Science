```python
# Importar la libreria

import pandas as pd
import numpy as np
```

## Series

```python
a = pd.Series(['Matem√°ticas', 'Historia', 'Econom√≠a', 'Programaci√≥n', 'Ingl√©s'], dtype='string')
print(a)
```

```python
b = pd.Series({'Matem√°ticas': 80, 'Historia': 80, 'Econom√≠a': 40, 'Programaci√≥n': 95, 'Ingl√©s': 85})
b
```

```python
#Atributos de una Serie
print("Tama√±o de la serie: ", b.size)
print("lista de nombres de las filas: ", b.index)
print("Tipo de datos",b.dtype)
```

### Acceso por √≠ndice

```python
b["Econom√≠a"]
print(b.dtype)
```

```python
b[['Historia', 'Econom√≠a', "Programaci√≥n"]]
```

```python
b[1:4]
```

### Resumen Descriptivo de una serie

```python
print("numero de elementos ",b.count())
print("Suma acumulada\n",b.cumsum()) # 50 + 75 = 125 | 125 +80=205
```

```python
# Serie con la frecuencia (n√∫mero de repeticiones) de cada valor n la serie b.
b.value_counts()
```

```python
# Varianza y la Desviaci√≥n t√≠pica
print("Varianza: ",b.var())
print("Desviaci√≥n t√≠pica: ",b.std())
```

### Aplicar operaciones

```python
print(b - 10)
```

```python
b.sort_values(ascending=True)
```

## DataFrame

```python
# Series

# lista de nombre carrera y promedio
nombre = ["Manuel", "Richard", "Andres", "Adriana", "Pedro", "Hernan"]
carrera = ["Psicologia", "Matematica", "Ing Biomedica", "Ing. Industrial", "Ing. Sistemas", "Ing. Comercial"]
promedio = [90, 20, 70, 98, 35, 60]
edad = [21,24,43,34, 25, 23]

# Diccionario de listas
dict = {'nombre': nombre, 'carrera': carrera, 'promedio': promedio, 'edad': edad} 

df = pd.DataFrame(dict)

df
#print(df)
```

```python
import matplotlib.pyplot as plt
#df.plot()

plt.title("Gr√°fico 2")

plt.xlabel("Nombre")
plt.ylabel("Promedio")
plt.plot(df['nombre'], df['promedio'], df['edad']) # x , y
plt.legend(['Promedio', 'edad'])
plt.show() 
```
